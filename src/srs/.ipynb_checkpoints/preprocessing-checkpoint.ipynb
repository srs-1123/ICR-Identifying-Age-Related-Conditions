# Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
import seaborn as sns
import random
from scipy.stats import uniform, randint

# Suppress any warnings
import warnings
warnings.filterwarnings('ignore')

# Sklearn
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold, GridSearchCV
from sklearn import metrics

# kNN Imputation
from sklearn.impute import KNNImputer

# Feature Selection
from sklearn.feature_selection import SelectKBest, f_classif

# Data Encoder and Scaler
import category_encoders as encoders
from sklearn.preprocessing import LabelEncoder, RobustScaler

# SMOTE (Synthetic Minority Over-sampling Technique)
from imblearn.over_sampling import SMOTE
```

```python
# データの読み込み
df_train = pd.read_csv("../data/train.csv")
df_test = pd.read_csv("../data/test.csv")
df_greeks = pd.read_csv("../data/greeks.csv")
# 訓練データから数値データのみ抽出
df_train_numerical = pd.read_csv("../data/train_numerical.csv")
# df_train_numerical = train.drop(['Id', 'EJ', 'Class'], axis=1)
# df_train_numerical.to_csv('../data/train_numerical.csv', index=False)
```

```python
# データフレームを結合
merged_data = pd.merge(df_train, df_greeks, on = 'Id', how='left')
merged_data.head()
```

## 1. 欠損値処理
kNN(k-nearest neighbors)インピュータは、データセット中の欠損値を、そのk個の最近傍の値に基づいて推定することによって埋める手法である。サンプル間の類似性を利用して欠損値を埋めるため、不完全なデータを扱うのに有用。

```python
# インスタンス生成
imputer = KNNImputer(n_neighbors=5)
# 訓練データに欠損値代入
df_train_imputed = pd.DataFrame(imputer.fit_transform(df_train[df_train_numerical.columns]), columns=df_train_numerical.columns)
# テストデータに欠損値代入
df_test_imputed = pd.DataFrame(imputer.transform(df_test[df_train_numerical.columns]), columns=df_train_numerical.columns)
```

```python
# 他にも欠損値があるかどうか確認
df_train_null = df_train_imputed[df_train_imputed.isnull().any(axis=1)]
df_test_null = df_test_imputed[df_test_imputed.isnull().any(axis=1)]

# 欠損値の数を出力
print('No. of records with missing value in Train data set after Imputation : {}'.format(df_train_null.shape[0]))
print('No. of records with missing value in Test data set after Imputation: {}'.format(df_test_null.shape[0]))
```

```python
# 元の訓練データも欠損値を補完したデータに置き換える
df_train_2 = df_train.drop(df_train_numerical.columns, axis=1)
df_train_2 = pd.concat([df_train_2, df_train_imputed], axis=1)

# テストデータを欠損値を代入したデータに置き換える
df_test_2 = df_test.drop(df_train_numerical.columns, axis=1)
df_test_2 = pd.concat([df_test_2, df_test_imputed], axis=1)

# 訓練データとテストデータの要素数を確認
print('Shape of the Train data set: {}'.format(df_train_2.shape))
print('Shape of the Test data set: {}'.format(df_test_2.shape))
```

## 2. RobustScaler
RobustScalerは、Pythonのscikit-learnライブラリで利用可能なデータ前処理テクニック。ロバストな方法で数値データをスケーリングするために使用され、標準化や最小-最大スケーリングのような他のスケーリング方法と比較して、外れ値の存在に影響されにくい。

```python
# インスタンス生成
scaler = RobustScaler()

'''訓練データのスケーリング'''
# インデックスを抽出
index = df_train_2.index
# スケーリング
scaler_train = scaler.fit_transform(df_train_2[df_train_numerical.columns])
scaler_df_train = pd.DataFrame(scaler_train, columns=df_train_numerical.columns)
# インデックスを振りなおす
scaler_df_train.index = index
print('Shape of Scaled Train Data Set: {}'.format(scaler_df_train.shape))


'''テストデータのスケーリング'''
# インデックスを抽出
index = df_test_2.index
# スケーリング
scaler_test = scaler.fit_transform(df_test_2[df_train_numerical.columns])
scaler_df_test = pd.DataFrame(scaler_test, columns=df_train_numerical.columns)
# インデックスを振りなおす
scaler_df_test.index = index
print('Shape of Scaled Test Data Set: {}'.format(scaler_df_test.shape))
```

```python
# 元の訓練データもスケーリングしたデータに置き換える
df_train_2 = df_train_2.drop(df_train_numerical.columns, axis=1)
df_train_2 = pd.concat([df_train_2, scaler_df_train], axis=1)

# テストデータをスケーリングしたデータに置き換える
df_test_2 = df_test_2.drop(df_train_numerical.columns, axis=1)
df_test_2 = pd.concat([df_test_2, scaler_df_test], axis=1)

# 訓練データとテストデータの要素数を確認
print('Shape of the Train data set: {}'.format(df_train_2.shape))
print('Shape of the Test data set: {}'.format(df_test_2.shape))
```

## 3. カテゴリカルデータのエンコーディング
CATBoostEncode（CatBoostEncoder）は、CatBoostアルゴリズム専用に設計されたカテゴリカルエンコーディング技術である。CatBoostは勾配ブースティング・アルゴリズムであり、カテゴリカルな特徴を扱う際に強力な性能を発揮することで知られている。CatBoostEncoderは、CatBoostアルゴリズムと互換性があり、モデルのパフォーマンスを向上させるためにカテゴリ変数を効果的にエンコードすることを目的とした、特別なエンコーディング手法。

```python
# インスタンス生成
CATBoostENCODE = encoders.CatBoostEncoder()
# カテゴリーデータの列名を変数に代入
categorical_cols = ['EJ']

'''CatBoostをカテゴリデータのエンコーディングに使用'''
# 訓練データのエンコーディング
encoder_train = CATBoostENCODE.fit_transform(df_train[categorical_cols], df_train['Class'])
encoded_df_train = pd.DataFrame(encoder_train)
print('Shape of the Encoded Train Data Set: {}'.format(encoded_df_train.shape))

# テストデータのエンコーディング
encoder_test = CATBoostENCODE.transform(df_test[categorical_cols])
encoded_df_test = pd.DataFrame(encoder_test)
print('Shape of the Encoded Test Dat Set: {}'.format(encoded_df_test.shape))
```

```python
'''元のデータをエンコーディング後のデータに置き換える'''
# 訓練データ
df_train_2 = df_train_2.drop(categorical_cols, axis=1)
df_train_2 = pd.concat([df_train_2, encoded_df_train], axis=1)
# テストデータ
df_test_2 = df_test_2.drop(categorical_cols, axis=1)
df_test_2 = pd.concat([df_test_2, encoded_df_test], axis=1)
# 訓練データとテストデータの要素数を確認
print('Shape of the Train data set: {}'.format(df_train_2.shape))
print('Shape of the Test data set: {}'.format(df_test_2.shape))
```

## 4. 特徴量選択（SelectKBest）
SelectKBestはScikit-learn (sklearn)ライブラリの特徴選択メソッドで、統計的なテストスコアに基づいてデータセットから上位k個の特徴（列）を選択する。SelectKBestの基本的な考え方は、スコアリング関数を使用して各特徴の統計的有意性を評価し、最も高いスコアを持つ上位k個の特徴を選択すること。
### 補足（F値の計算方法について）
* SelectKBestで使用できるスコア関数の一つであるf_classifは、ANOVA（Analysis of Variance、分散分析）のF値を計算します。これはカテゴリカルな目標変数と連続値の特徴量との間の統計的な関連性を評価します。具体的な計算方法は以下の通りです。

1. 各クラス内での特徴量の値の分散（Within-group variance）を計算します。これは各クラス内での特徴量の値のばらつきを表します。

2. クラス間での特徴量の値の分散（Between-group variance）を計算します。これは各クラス間での特徴量の値のばらつきを表します。

3. F値はこれら二つの分散の比率として定義されます。具体的には、Between-group varianceをWithin-group varianceで除したものがF値になります。

* したがって、F値は各クラス内での特徴量の値のばらつきが小さく（つまり、同じクラスのサンプルに対する特徴量の値が似ている）、クラス間での特徴量の値のばらつきが大きい（つまり、異なるクラスのサンプルに対する特徴量の値が異なる）場合に大きくなります。その結果、目標変数との関連性が強い特徴量が選択されます。

* なお、F値とp値はANOVAの結果から計算されます。p値はF値が偶然によるものである確率を示し、p値が小さいほど特徴量が目標変数と統計的に有意に関連していると解釈できます。

```python
# オーバーサンプリングを行う前に、訓練データを説明変数と目的変数に分割
X_train = df_train_2.drop(['Id', 'Class'], axis=1)
y_train = df_train_2['Class']
# 列名を変更
y_train.columns = ['Class']
```

```python
'''初期値'''
# 有意水準
threshold = 0.1
# F値の閾値
f_score = 5
'''F値とp値を計算'''
# インスタンス生成
#     回帰: f_regression, mutual_info_regression
#     分類: chi2, f_classif(分散分析のF値), mutual_info_classif
# この時点ではkをもとの訓練データと同じにする
fs = SelectKBest(score_func=f_classif, k=len(X_train.columns))
# 特徴量選択
X_selected = fs.fit_transform(X_train, y_train.values)
print('Before the SelectKBest = {}'.format(X_train.shape))

'''選択したF値とp値と設定した閾値を用いて特徴量を選択'''
new_features = [] # 選択された特徴量を格納
drop_features = [] # 使わない特徴量を格納

# F値が大きく、p値の小さい特徴量を選択
for i in range(len(X_train.columns)):
    print('Feature {}: {:.3f} with p-value {:.3f}'.format(X_train.columns[i], fs.scores_[i], fs.pvalues_[i]))
    if fs.pvalues_[i] <= threshold and fs.scores_[i] >= f_score:
        new_features.append(X_train.columns[i])
    else:
        drop_features.append(X_train.columns[i])
        
# 訓練データから選択した特徴量を抽出        
X_selected_final = pd.DataFrame(X_selected)
X_selected_final.columns = X_train.columns
X_selected_final = X_selected_final[new_features]
print('=' * 30)
print('After the SelectKBest = {}'.format(X_selected_final.shape))
print('Drop-out Features = {}'.format(len(drop_features)))
```

```python
# 元のデータに反映
X_train = X_train.drop(drop_features, axis=1)
df_train_2 = df_train_2.drop(drop_features, axis=1)
df_test_2 = df_test_2.drop(drop_features, axis=1)
```

## 5. オーバーサンプリング（SMOTE） 
SMOTE (Synthetic Minority Over-sampling Technique) は、不均衡なデータセットの問題に対処する手法である。不均衡なデータセットとは、対象変数のクラスが均等に表現されておらず、分布が歪んでいる場合に発生する。これは、詐欺検出、希少疾患予測、異常検出など、多くの実世界の分類問題でよく見られる問題である。

```python
# 訓練データと検証データに分割
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3)

print('Shape of train : {}'.format(X_train.shape))
print('Shape of test : {}'.format(X_test.shape))
print('='*50)
print('Shape of df_train (incl. ID and Class): {}'.format(df_train_2.shape))
print('Shape of df_test (incl. ID): {}'.format(df_test_2.shape))
```

```python
# インスタンス生成
smote = SMOTE(sampling_strategy={0: 360, 1: 360})
# オーバーサンプリング
X_smote, y_smote = smote.fit_resample(X_train, y_train)

# 結果を出力
print("元のデータ数",len(df_train_2))
print("元のデータの陽性率: {:.2%}".format(len(y_train[y_train==1])/len(y_train)))
print("元のデータの陰性率: {:.2%}".format(len(y_train[y_train==0])/len(y_train)))
print("オーバーサンプル後のデータ数",len(X_smote))
print("元のデータの陽性率: {:.2%}".format(len(y_smote[y_smote ==1])/len(y_smote)))
print("元のデータの陰性率: {:.2%}".format(len(y_smote[y_smote ==0])/len(y_smote)))
```
